# Azure Application Insights

## What is Azure Application Insights?
Azure Application Insights is an **Application Performance Monitoring (APM)** service in Azure that helps developers monitor the **performance, availability, and health** of their applications in real time.

---

## Why is it used?
It helps identify:
- Slow API requests
- Application crashes and exceptions
- Performance bottlenecks
- User activity and usage
- External service failures (database, APIs, etc.)

This allows developers to quickly diagnose and fix issues in production.

---

## What does it monitor?
- **Requests** – Incoming API calls
- **Dependencies** – Calls to databases, APIs, Azure services
- **Exceptions** – Errors and stack traces
- **Performance** – Response time, CPU, memory
- **Logs** – Custom application logs
- **Live Metrics** – Real-time application monitoring

A LangGraph application performs the following steps:

```text
User Question
      │
      ▼
Retriever
      │
      ▼
LLM
      │
      ▼
Final Answer
```

LangSmith traces every step:

```text
Run
├── User Input
├── Retriever
│     ├── Retrieved Documents
│     └── Retrieval Time
├── LLM
│     ├── Prompt
│     ├── Tokens Used
│     ├── Response
│     └── Latency
└── Final Output
```

This makes it easy to understand where errors or delays occur.

---

## Why is it important?

Imagine your chatbot gives a wrong answer.

Without LangSmith:
- You only see the incorrect output.
- It is difficult to know why it happened.

With LangSmith:
- You can inspect the prompt.
- See which documents were retrieved.
- Check tool calls.
- Measure execution time.
- Identify the exact source of the problem.

---

## Advantages

- End-to-end execution tracing
- Easy debugging of AI workflows
- Prompt evaluation and comparison
- Performance monitoring
- Dataset creation for evaluation
- Team collaboration through shared traces

---

## Summary

LangSmith is an **observability and debugging platform for LLM applications**. It records every step of an AI application's execution, allowing developers to monitor performance, debug errors, evaluate model responses, and improve the overall quality of AI systems.
